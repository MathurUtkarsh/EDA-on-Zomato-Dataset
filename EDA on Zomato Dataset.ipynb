```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When the rating is between 4.5 to 4.9 ---> Excellent
2. When the ratings is between 4.0 to 3.4 ---> Very Good
3. When the rating is between 3.5 to 3.9 ---> Good
4. When the rating is between 3.0 to 3.4 ---> Average
5. When the rating is between 2.5 to 2.9 ---> Average
6. When the rating is between 2.0 to 2.4 ---> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette = ['blue','red','orange','yellow','green','green'])
```

## Observations:
1. Not rated count is very high.
2. Maximum number of rating is between 2.5 to 3.4

```python
sns.countplot(x='Rating color', data=ratings, palette = ['blue','red','orange','yellow','green','green'])
# It is not rating count, It is frequency of color.
```
